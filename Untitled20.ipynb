## Accessing Slices of DataFrame

loc[star_row:end_row:start_col:end_col]

iloc[star_row_pos:end_row_pos:start_col_pos:end_col_pos]
> Add blockquote



```python
data={
    'name':['lucky','ehtesham','azim','imran'],
    'marks':[80,81,60,33],
    'age':[21,22,23,24]
    }
df=pd.DataFrame(data,index=['A','B','C','D'])
df
```

```python
df.loc['B':,:'age']
```

```python
df.iloc[1:,:2]
```

```python
df.iloc[[2,3],[1]]
```

```python
df.loc[['C','A'],['age']]
```

# Deleting Columns by 3 ways
